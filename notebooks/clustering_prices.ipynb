This file attempts to create price clusters using k-means and agglomerative clustering.

```python
import pandas as pd
import numpy as np
from  matplotlib import pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
# Pull the training data
data_train = pd.read_csv('../data/data_train.csv')

y_train = data_train['SalePrice']
```

```python
# Cluster using k-means
n_clusters = [3,4,5]

for k in n_clusters:
    cluster = KMeans(n_clusters=k)
    cluster.fit(y_train.values.reshape(-1,1))
    
    print('k =', k, ':')
    for i in range(k):
        print('Min value of cluster', i+1, 'is: ', y_train[cluster.labels_==i].min())
        print('Max value of cluster', i+1, 'is: ', y_train[cluster.labels_==i].max())
    
    colormap = cm.rainbow(np.linspace(0, 1, len(cluster.cluster_centers_)))
    plt.scatter(y_train.index, y_train, c=colormap[cluster.labels_])
    plt.title('k = ' + str(k))
    plt.show()
```

```python
# Cluster using agglomerative clustering
n_clusters = [3,4,5]

for k in n_clusters:
    cluster = AgglomerativeClustering(n_clusters=k)
    cluster.fit(y_train.values.reshape(-1,1))
    
    print('k =', k, ':')
    for i in range(k):
        print('Min value of cluster', i+1, 'is: ', y_train[cluster.labels_==i].min())
        print('Max value of cluster', i+1, 'is: ', y_train[cluster.labels_==i].max())
```
