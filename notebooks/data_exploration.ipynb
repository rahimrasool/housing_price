This file explores the cleaned dataset. It checks that there is no missing data and plots sales price as a function of various quantitative variables.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from mpl_toolkits import mplot3d
```

```python
# Import cleaned dataset
data = pd.read_csv('../data/data_train.csv')
```

```python
# Count missing values
pd.set_option('display.max_rows', 400)
data.isna().sum()
```

```python
# Print descriptive statistics
pd.set_option('display.max_columns', None)
data.describe()
```

```python
# Plot distribution of sales price
plot = sns.histplot(data, x="SalePrice")
plot.set_title('Distribution of Sale Prices')
plot.set(xlabel='Sale Price (dollars)', ylabel='Count')
```

```python
# Select variables
target = data['SalePrice']
features = data.drop(['SalePrice'], axis=1)


# K-means clustering model
n_clusters = 3

cluster = KMeans(n_clusters=n_clusters, random_state=2)
pred_clusters = cluster.fit(features).labels_

# Plot sale price
colormap = cm.rainbow(np.linspace(0, 1, n_clusters))
plt.scatter(target.index, target, c=colormap[cluster.labels_])
plt.title('Clusters')
plt.ylabel('Sale Price')
plt.xlabel('Data Point Index')
plt.show()

# Plot features in three dimensions
pca = PCA(n_components=3)
pca_features = pca.fit_transform(features)

fig = plt.figure(figsize=(16,16))
ax = plt.axes(projection='3d')
ax.scatter3D(pca_features[cluster.labels_==0, 0], pca_features[cluster.labels_==0, 1],
             pca_features[cluster.labels_==0, 2], cmap=cm.rainbow(0));
ax.scatter3D(pca_features[cluster.labels_==1, 0], pca_features[cluster.labels_==1, 1],
             pca_features[cluster.labels_==1, 2], cmap=cm.rainbow(0.5));
ax.scatter3D(pca_features[cluster.labels_==2, 0], pca_features[cluster.labels_==2, 1],
             pca_features[cluster.labels_==2, 2], cmap=cm.rainbow(1));
ax.set_xlabel('First Principle Component')
ax.set_ylabel('Second Principle Component')
ax.set_zlabel('Third Principle Component')

# Print descriptive statistics
cluster_labels=[[] for i in range(n_clusters)]
for i, j in enumerate(pred_clusters):
    cluster_labels[j].append(target.index[i])
for i in range(n_clusters):
    print('Cluster ', i + 1, ':')
    c = target.loc[cluster_labels[i]]
    print(c.describe())
```

```python
# Agglomerative clustering model
n_clusters = 3

agglom_cluster = AgglomerativeClustering(n_clusters=n_clusters)
pred_clusters = agglom_cluster.fit(features).labels_

# Plot sales price
plt.scatter(target.index, target, c=colormap[agglom_cluster.labels_])
plt.title('Clusters')
plt.ylabel('Sale Price')
plt.xlabel('Data Point Index')
plt.show()

# Plot features in three dimensions
pca = PCA(n_components=3)
pca_features = pca.fit_transform(features)

fig = plt.figure(figsize=(16,16))
ax = plt.axes(projection='3d')
ax.scatter3D(pca_features[agglom_cluster.labels_==0, 0], pca_features[agglom_cluster.labels_==0, 1],
             pca_features[agglom_cluster.labels_==0, 2], cmap=cm.rainbow(0));
ax.scatter3D(pca_features[agglom_cluster.labels_==1, 0], pca_features[agglom_cluster.labels_==1, 1],
             pca_features[agglom_cluster.labels_==1, 2], cmap=cm.rainbow(0.5));
ax.scatter3D(pca_features[agglom_cluster.labels_==2, 0], pca_features[agglom_cluster.labels_==2, 1],
             pca_features[agglom_cluster.labels_==2, 2], cmap=cm.rainbow(1));
ax.set_xlabel('First Principle Component')
ax.set_ylabel('Second Principle Component')
ax.set_zlabel('Third Principle Component')

# Print descriptive statistics
cluster_labels=[[] for i in range(n_clusters)]
for i, j in enumerate(pred_clusters):
    cluster_labels[j].append(target.index[i])
for i in range(n_clusters):
    print('Cluster ', i + 1, ':')
    c = target.loc[cluster_labels[i]]
    print(c.describe())
```
